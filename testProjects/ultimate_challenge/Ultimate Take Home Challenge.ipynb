# Part 1 ‐ Exploratory data analysis
The attached logins.json file contains (simulated) timestamps of user logins in a particular geographic location. Aggregate these login counts based on 15­minute time intervals, and visualize and describe the resulting time series of login counts in ways that best characterize the underlying patterns of the demand. Please report/illustrate important features of the demand, such as daily cycles. If there are data quality issues, please report them.

```python
# Data wrangling
import pandas as pd
import numpy as np
import json
from pandas.io.json import json_normalize

# Data visualization
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(font_scale=1.5)

from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15, 6

# Machine learning models
from sklearn.preprocessing import normalize
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score
```

```python
logins=pd.read_json('logins.json')
logins['count']=1
logins.set_index('login_time',inplace=True)
logins.info()
logins.head()
```

**Summary**<br>
The logins dataset contains login records from 1970-01-01 to 1970-04-13. The 'login_time' is not in chronological order and needs to be aggregated to gain insights. We've added the column 'count' for aggregation purpose.

```python
#Aggregate these login counts based on 15-minute time intervals
logins.resample('15T').sum().plot(legend=False)
plt.ylabel('Login Counts')
plt.title('15 Mins Interval');

```

```python
#Aggregate these login counts on a monthly basis. Note April has data till April 13.
logins.resample('MS').sum().plot(kind='bar',legend=False,rot=0)
plt.ylabel('Login Counts')
plt.xticks(np.arange(0,4),(['Jan','Feb','Mar','Apr']))
plt.title('Monthly Interval');
```

```python
# Extract the hour from the index of logins
idx_hour=logins.index.hour

# Aggregate the logins by hour
login_by_hour=logins.groupby(idx_hour).sum(axis=1)

# Visualize the hourly logins
ax=login_by_hour.plot(kind='bar',legend=False,rot=0)

plt.xticks(np.arange(0,24),([str(i) for i in range(24)]))
plt.xlabel('Hour_of_Day')
plt.xlim(-0.5,23.5)
plt.ylabel('Login Counts')
plt.title('Total Logins v.s. Hour');
```

```python
# Extract the day from the index of logins
import calendar
weekday=list(calendar.day_name)
idx_wday=logins.index.weekday_name

# Aggregate the logins by day
login_by_wday=logins.groupby(idx_wday).sum(axis=1).reindex(weekday)

# Visualize the daily logins
login_by_wday.plot(kind='bar',legend=False,rot=0)
plt.xticks(np.arange(0,7),login_by_wday.index)
plt.xlabel('Day_of_Week')
plt.ylabel('Login Counts')
plt.title('Total Logins v.s. Day of Week');
```

```python
logins['hour']=idx_hour
logins['wday']=idx_wday

sns.countplot(x='hour',hue='wday',data=logins,hue_order=weekday,palette='Spectral');
plt.ylabel('Login Counts')
plt.title('Total Logins v.s. Hour and Wday');
```

**Summary**<br>
* The '15 Mins Interval' plot displays some periodic oscillation over time but it's very difficult to extract information directly there.<br>
* The 'Monthly Interval' plot indicates an increasing trend in login activities from January to March and very promisingly April since the April data till April 13th. More data is needed to confirm the trend.<br>
* The 'Total Logins v.s. Hour' plot shows that there are spikes around lunch breaks and during night on a daily basis.<br>
* The 'Total Logins v.s. Day of Week' plot shows an increasing trend of logins in a week starting from Monday. Weekends have more logins than weekdays.<br>
* The last plot 'Total Logins v.s. Hour and Wday' suggests there are two types of logins preferences. Weekend users are very active after mid-night on weekends and in the afternoon during the day. Weekday users are very active around lunch breaks and in the evening hours. 

# Part 2 ‐ Experiment and metrics design
The neighboring cities of Gotham and Metropolis have complementary circadian rhythms: on weekdays, Ultimate Gotham is most active at night, and Ultimate Metropolis is most active during the day. On weekends, there is reasonable activity in both cities.

However, a toll bridge, with a two-way toll, between the two cities causes driver partners to tend to be exclusive to each city. The Ultimate managers of city operations for the two cities have proposed an experiment to encourage driver partners to be available in both cities, by reimbursing all toll costs.

1. What would you choose as the key measure of success of this experiment in encouraging driver partners to serve both cities, and why would you choose this metric?

2. Describe a practical experiment you would design to compare the effectiveness of the proposed change in relation to the key measure of success. Please provide details on:<br>
    a. how you will implement the experiment<br>
    b. what statistical test(s) you will conduct to verify the significance of the observation<br>
    c. how you would interpret the results and provide recommendations to the city operations team along with any caveats.<br>

**Solution**<br>
I would suggest the increase in the number of toll reimbursement for both directions between cities as the key measure of success as it indicates that the plan does encourage drivers to cross cities.

I will divide data into two groups: a control group without reimbursement and an experimental group with reimbursement. Then we can use z-test or t-test depending on sample size to check if the average number of reimbursements is different between the two groups. The null hypothesis is that there is no difference in the average number of reimbursements. If the resulting test yields a p-value less than 0.05, we reject the null hypothesis and admit there is significant difference and hence we confirm that reimbursing toll costs works. If the p-value is larger than 0.05, then we can't reject the null hypothesis. We need to check other factors such as drivers' behavior, relevant features comparison between two cities etc.

# Part 3 ‐ Predictive modeling
Ultimate is interested in predicting rider retention. To help explore this question, we have provided a sample dataset of a cohort of users who signed up for an Ultimate account in January 2014. The data ('ultimate_data_challenge.json') was pulled several months later; we consider a user retained if they were “active” (i.e. took a trip) in the preceding 30 days.

We would like you to use this data set to help understand what factors are the best predictors for retention, and offer suggestions to operationalize those insights to help Ultimate. The data is in the attached file ultimate_data_challenge.json. 

Data description<br>
* city: city this user signed up in
* phone: primary device for this user
* signup_date: date of account registration; in the form ‘YYYYMMDD’
* last_trip_date: the last time this user completed a trip; in the form ‘YYYYMMDD’
* avg_dist: the average distance in miles per trip taken in the first 30 days after signup
* avg_rating_by_driver: the rider’s average rating over all of their trips
* avg_rating_of_driver: the rider’s average rating of their drivers over all of their trips
* surge_pct: the percent of trips taken with surge multiplier > 1
* avg_surge: The average surge multiplier over all of this user’s trips
* trips_in_first_30_days: the number of trips this user took in the first 30 days after signing up
* ultimate_black_user: TRUE if the user took an Ultimate Black in their first 30 days; FALSE otherwise
* weekday_pct: the percent of the user’s trips occurring during a weekday

## Load data

```python
# Load json
json_file=json.load(open('./ultimate_data_challenge.json'))

# Flattern json
data=json_normalize(json_file)
data.info()
data.head()
```

## Data wrangling 
We proceed with data wrangling in the following steps:
* Find and remove duplicate entries
* Check and handle missing values 
    * Drop a small number of entries without phones
    * Fill missing values with its respective mean for 'avg_rating_by_driver' and 'avg_rating_of_driver'
* Convert date features to datetime type<br>
* Add target feature 'retained_user' indicating whether users were “active” (i.e. took a trip) in the 30 days before pulling out the data. 

### Find and drop duplicates

```python
print('Number of duplicated observations: ', data.duplicated().sum())
```

```python
data.drop_duplicates(keep='first',inplace=True)
```

### Missing values

```python
print('Percentage of missing values (%): ')
(data.isnull().sum()/data.shape[0]*100).round(1)
```

Features with missing values include: 'avg_rating_by_driver'(0.4%), 'avg_rating_of_driver' (16.2%) and 'phone' (0.8%). So we drop the phone missing entries since it takes only less than 1%. We fill missing values for the other two features with their respective means.

```python
# Handle missing values
data.dropna(subset=['phone'],how='all',inplace=True)
values={'avg_rating_by_driver': np.round(data.avg_rating_by_driver.mean(),1),
        'avg_rating_of_driver': np.round(data.avg_rating_of_driver.mean(),1)}
data.fillna(value=values,inplace=True)
```

### Datetime

```python
#Format datetime columns
data.signup_date=pd.to_datetime(data.signup_date)
data.last_trip_date=pd.to_datetime(data.last_trip_date)
```

### Target feature 'retained_user'

```python
#we consider a user retained if they were “active” (i.e. took a trip) in the preceding 30 days before pulling out.
from datetime import datetime, timedelta

pullout_date=data.last_trip_date.max()
print('The data pull_out date is: ', pullout_date.date())

cutoff_date=(pullout_date-timedelta(days=30)).date()
print('Cutoff date: ',cutoff_date)

data['retained_user']=np.where(data.last_trip_date>=cutoff_date,1,0)
print('non retained user: 0, retained user: 1')
print(data.retained_user.value_counts())
```

## Build Predictive Models
The aim is to build a predictive model to help Ultimate determine whether or not a user will be active in their 6th month on the system. I choose random forest classifier to build the model as it fits an ensemble of trees and use averaging to improve the predictive accuracy and control over-fitting. So it's a good choice for this classification problem. In addition, we can gain insights on important features via its feature_importances_ attribute. The random forest classifier has lots of parameters to tune. It takes time to find the optimal parameters. This is an imbalanced binary classification. It might be hard to predict the minor class, i.e., 'retained_user'. 

```python
# Perform one-hot encoding of categorical features
data=pd.get_dummies(data=data,columns=['city','phone','ultimate_black_user'], drop_first=True)
data.reset_index(drop=True,inplace=True)
data.info()
```

```python
X=normalize(data.drop(columns=['signup_date','retained_user','last_trip_date'],axis=1))
y=data.retained_user

X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.75,random_state=100)
```

```python
rfc=RandomForestClassifier()
rfc.fit(X_train,y_train)
y_pred=rfc.predict(X_test)

print ("Accuracy : {:.0f}%".format(accuracy_score(y_test,y_pred)*100))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Report :\n", classification_report(y_test,y_pred))
```

The average performance (accuracy, precision, recall and f1-score) is good. Collecting more data for retained users or considering oversampling or undersampling of the data might improve the model performance. Let's find the most important features selected by the model below.

```python
importance=rfc.feature_importances_
features=data.drop(columns=['signup_date','retained_user','last_trip_date'],axis=1).columns.values

df=pd.DataFrame({'Features': features, 'Importance': rfc.feature_importances_}).sort_values('Importance',ascending=True).set_index('Features')
df.plot(kind='barh',figsize=(12,10))
plt.title('Important Features');
```
